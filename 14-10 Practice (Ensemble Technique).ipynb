```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a=pd.read_csv("C:\All Datasets\Brain Stroke.csv")
a
```

```python
a.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
a["gender"]=l.fit_transform(a["gender"])
a["ever_married"]=l.fit_transform(a["ever_married"])
a["work_type"]=l.fit_transform(a["work_type"])
a["Residence_type"]=l.fit_transform(a["Residence_type"])
a["smoking_status"]=l.fit_transform(a["smoking_status"])
```

```python
a
```

```python
x=a.iloc[:,:-1].values
y=a.iloc[:,-1].values
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
x1=s.fit_transform(x)
```

```python
from imblearn.under_sampling import RandomUnderSampler
r=RandomUnderSampler()
X,Y=r.fit_resample(x1,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
l1=LogisticRegression()
sv=SVC(kernel="linear")
rf=RandomForestClassifier()
```

```python
l1.fit(x_train,y_train)
sv.fit(x_train,y_train)
rf.fit(x_train,y_train)
```

```python
p=l1.predict(x_train)
p1=sv.predict(x_train)
p2=rf.predict(x_train)
```

```python
print("Logistic",p,"SVM",p1,"RandomForest",p2)
```

```python
from sklearn.ensemble import VotingClassifier
vc=VotingClassifier(estimators=[("logistic",l1),("SVMo",sv),("RandomF",rf)])
vc.fit(x_train,y_train)
```

```python
vc.predict(x_train)
```
